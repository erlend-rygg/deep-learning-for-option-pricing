```python
import pandas as pd
heston_filenames = [
    '2015-01-02_2015-12-31',
    '2016-01-04_2016-12-30',
    '2017-01-03_2017-12-29',
    '2018-01-02_2018-12-31',
    '2019-01-02_2019-12-31',
    '2020-01-02_2020-12-31',
    '2021-01-04_2021-12-31',
    '2022-01-03_2022-12-30'
]
```

```python
# Read all files and concatenate them into one dataframe
df_r = pd.DataFrame()
for filename in heston_filenames:
    df_r = pd.concat([df_r, pd.read_csv(f'../data/results/Heston/Results/{filename} Heston results.csv')])

df_r.info()
print(len(df_r))
```

```python
# Read all files and concatenate them into one dataframe
df_p = pd.DataFrame()
for filename in heston_filenames:
    df_p = pd.concat([df_p, pd.read_csv(f'../data/results/Heston/Parameters/{filename} Heston parameters.csv')])

df_p.info()
print(len(df_p))
```

```python
parameters = ['sigma', 'kappa', 'theta', 'v0' , 'rho']

for parameter in parameters:
    df_p['Quote_date'] = pd.to_datetime(df_p['Quote_date'])
    df_p.plot(x='Quote_date', y=parameter, figsize=(15, 5), title=f'{parameter} parameter')
```
